```python

import pandas as pd
from sqlalchemy import create_engine 
import pymysql.cursors
import os
# import getpass
import urllib.parse
```

```python
pw_raw = os.getenv('sql_password')
pw = urllib.parse.quote_plus(pw_raw)
```

```python
df = pd.read_csv("C:/Users/darce/Ironhack/week6/mini_api_project/comptage-velo-donnees-compteurs.csv", encoding="ISO-8859-1")
```

```python
bd = "paris_bike_traffic"
connection_string = 'mysql+pymysql://root:' + pw + '@localhost/'+bd
engine = create_engine(connection_string)
engine
```

```python
# Create a SQLAlchemy engine object to connect to your SQL database
# Replace 'database_type' with your database type (e.g., 'mysql', 'postgresql', 'sqlite')
# Replace 'username', 'password', 'hostname', and 'database_name' with your database credentials
# engine = create_engine('mysql://root:pw@localhost3306/paris_bike_traffic')

# Assuming 'df' is your Pandas DataFrame containing the data
# Replace 'table_name' with the name you want to give to the SQL table
df.to_sql('donnes_compteurs', con=engine, schema='paris_bike_traffic', if_exists='replace', index=False)

# 'if_exists' parameter specifies what to do if the table already exists
# 'replace' will replace the existing table with the DataFrame data
# 'index=False' will not include the DataFrame index as a separate column in the SQL table
```

```python
excel_file = pd.ExcelFile("accidents_velo_dim_tables.xlsx")

for sheet_name in excel_file.sheet_names:
    # Read the sheet into a DataFrame
    df4 = excel_file.parse(sheet_name)

    # Export the DataFrame to SQL
    df4.to_sql(sheet_name, con=engine, if_exists='replace', index=False)

    print(f"Exported {sheet_name} to SQL successfully.")
```
